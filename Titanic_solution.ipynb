```python
data_train = pd.read_csv("/kaggle/input/titanic/train.csv")
data_test = pd.read_csv("/kaggle/input/titanic/test.csv")
data_train.head()
```

```python
train_features = data_train[['Pclass', 'Sex', 'Age', 'SibSp','Parch', 'Fare','Embarked']]
lables = data_train["Survived"]
test_data = data_test[['Pclass', 'Sex', 'Age', 'SibSp','Parch', 'Fare','Embarked']]
```

```python

#Converting to boolean to use as a feature for training
train_features['Sex'] = train_features['Sex']=='male'
train_features["Age"] = train_features["Age"].fillna(train_features["Age"].mean())
train_features["Embarked"] = pd.get_dummies(train_features["Embarked"])
# train_features['Fare'] = (train_features.Fare - train_features.Fare.std())/train_features.Fare.mean()
#same feature chages for test set
test_data['Sex'] = test_data['Sex']=='male'
test_data["Age"] = test_data["Age"].fillna(test_data["Age"].mean())
test_data.Fare.fillna(test_data.Fare.mean(),inplace = True)
test_data["Embarked"] = pd.get_dummies(test_data["Embarked"])
```

```python
drop_index = data_train[train_features.Fare>300].index
train_features.drop(drop_index , inplace = True)
lables.drop(drop_index, inplace=True)
```

```python
# x_train, x_test, ytrain, y_test = train_test_split(train_features, lables,test_size = 0.33, random_state=42)
model_gb = RandomForestClassifier(n_estimators=100, max_depth=5)
model_gb.fit(train_features,lables)
predictions = model_gb.predict(test_data)
# print(accuracy_score(predictions,y_test))
output = pd.DataFrame({'PassengerId': data_test.PassengerId, 'Survived': predictions})
```

```python
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
